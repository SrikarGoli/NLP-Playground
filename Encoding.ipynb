## OneHot encoding

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
text="I love NLP. I love AI"
tokens=word_tokenize(text.lower())
tokens=[word for word in tokens if word.isalpha()]
vocabulary=list(set(tokens))
print("vocabulary: ",vocabulary)
```

```python
word_idx={word:idx for idx,word in enumerate(vocabulary)}
one_hot_encoded=[]
for word in vocabulary:
    encode=[0]*len(vocabulary)
    encode[word_idx[word]]=1
    one_hot_encoded.append(encode)

for i,j in zip(vocabulary,one_hot_encoded):
    print("word ",i,"one hot encoding",j)
```

## Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
sentences=["The cat sat on the mat","The cat movie is good","The cat movie is scary as it is"]

vectorizer=CountVectorizer()
bow_matrix=vectorizer.fit_transform(sentences)
vocabulary=vectorizer.get_feature_names_out()
print(vocabulary)
bow_array=bow_matrix.toarray()
print("BOW: ",bow_array)
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
sentences=["The cat sat on the mat","The cat movie is good","The cat movie is scary as it is"]

vectorizer=TfidfVectorizer(stop_words='english')
tfidf_matrix=vectorizer.fit_transform(sentences)
vocabulary=vectorizer.get_feature_names_out()
print(vocabulary)
tfidf_array=tfidf_matrix.toarray()
print("Tfidf: ",tfidf_array)
```
